```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
num_points = 1000
vectors_set = []
for i in range(num_points):
    x1 = np.random.normal(0.0, 0.55)
    y1 = x1 * 0.1 + 0.3 + np.random.normal(0.0, 0.03)
    vectors_set.append([x1, y1])
x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]
plt.scatter(x_data, y_data, c='r')
plt.show()
```

```python
W = tf.Variable(tf.random.uniform([1], -1.0, 1.0), name='W')
b = tf.Variable(tf.zeros([1]), name='b')

loss_fn = lambda: tf.reduce_mean(tf.square((W * x_data + b)-y_data))
optimizer = tf.optimizers.SGD(0.5) #learn step, normally use 0.01

trainable_vars = [W,b]
epochs = 20
for _ in range(epochs):
    with tf.GradientTape() as tape:
        loss = loss_fn()
    gradients = tape.gradient(loss, trainable_vars)
    #optimizer.apply_gradients(zip(gradients, trainable_vars))
    W.assign_sub(gradients[0]*0.5)
    b.assign_sub(gradients[1]*0.5)
    print(loss)
    
print(W)
print(b)

```
